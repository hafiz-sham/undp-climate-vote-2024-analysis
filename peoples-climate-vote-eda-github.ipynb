```python
# "Age <18
# Sort by "Country"
#"Response" == "Very quickly" or "Somewhat quickly"
df.groupby("Country")["Weighted Mean"].max().reset_index()
```

```python
# 18-35 = young
young_df = df[
    (df["Age"] == "18 to 35") &
    (df["Response"].isin(["Very quickly", "Somewhat quickly"]))
]

young_df_result = (
    young_df
    .sort_values("Country")
    .groupby("Country", as_index=False)["Weighted Mean"]
    .max()
    .dropna(subset=["Weighted Mean"])
    .sort_values("Weighted Mean", ascending=False)
)

print(young_df_result)
```

```python
plt.figure(figsize=(12, 8))

sns.barplot(
    data=young_df_result, 
    x="Weighted Mean", 
    y="Country"
)

plt.title('Ages 18-35 who agreed it should be done quickly', fontsize=16)
plt.xlabel('% of people', fontsize=12)
plt.ylabel('Country', fontsize=12)



plt.tight_layout()

plt.show()
```

```python
df_renewables_age = df[df['Question Text (Short)'] == 'Renewable energy transition']
df_renewables_age = df_renewables_age[['Country', 'Response', 'Age', 'Weighted Mean']]

display(df_renewables_age)
```

```python
df_renewables_age_urgency = df_renewables_age[
    df_renewables_age['Response'].isin(['Very quickly', 'Somewhat quickly']) &
    (df_renewables_age['Age'] != 'All Ages')
].dropna(subset=['Weighted Mean'])

df_pivot_responses_age = df_renewables_age_urgency.pivot_table(
    index=['Country', 'Age'],
    columns='Response',
    values='Weighted Mean'
).fillna(0)

df_pivot_responses_age['Total_Urgency'] = df_pivot_responses_age['Very quickly'] + df_pivot_responses_age['Somewhat quickly']

age_pivot_total_urgency = df_pivot_responses_age.reset_index().pivot_table(
    index='Country',
    columns='Age',
    values='Total_Urgency'
)

age_order = ['Under 18', '18 to 35', '36 to 59', '60 plus']
existing_age_columns = [col for col in age_order if col in age_pivot_total_urgency.columns]
age_pivot = age_pivot_total_urgency[existing_age_columns]

display(age_pivot)
```

```python
final_age_columns = age_pivot.drop(columns=['Under 18'])
display(final_age_columns.head())
```

```python
# Drop rows that contain any NaN values in the table
# Keep a copy of the cleaned table and show before/after shapes
before_shape = final_age_columns.shape
final_age_columns_clean = final_age_columns.dropna(how='any').copy()
after_shape = final_age_columns_clean.shape
print(f"Shape before dropping NaNs: {before_shape}")
print(f"Shape after  dropping NaNs: {after_shape}")

# Optionally overwrite the original variable if you want to continue using the same name
# final_age_columns = final_age_columns_clean

# Show the cleaned data
display(final_age_columns_clean)
```

```python
# Remove any 'Global' row from the cleaned final_age_columns_clean table
# Print shapes before/after to confirm the removal
before_shape = final_age_columns_clean.shape

# If 'Country' is a column, filter that; otherwise drop by index (errors='ignore' to be safe)
if 'Country' in final_age_columns_clean.columns:
    final_age_columns_clean = final_age_columns_clean[final_age_columns_clean['Country'] != 'Global'].copy()
else:
    final_age_columns_clean = final_age_columns_clean.drop(index='Global', errors='ignore').copy()

after_shape = final_age_columns_clean.shape
print(f"Shape before removing 'Global': {before_shape}")
print(f"Shape after  removing 'Global': {after_shape}")

# Show the cleaned top rows
display(final_age_columns_clean)

# If you want to replace the original variable too, uncomment the next line:
# final_age_columns = final_age_columns_clean
```

```python
df_renewables_transition = df[
    (df['Question Text (Short)'] == 'Renewable energy transition') &
    (df['Response'].isin(['Very quickly', 'Somewhat quickly'])) &
    (df['Education'] == 'All Education') &
    (df['Age'] == 'All Ages')
]
df_renewables_transition_filtered_cols = df_renewables_transition[['Country', 'Response', 'Weighted Mean']]
display(df_renewables_transition_filtered_cols.head())
```

```python
df_combined_urgency_viz = df_renewables_transition_filtered_cols[
    df_renewables_transition_filtered_cols['Country'] != 'Global'
].groupby('Country')['Weighted Mean'].sum().reset_index()

# Sort the results from highest to lowest combined urgency
df_combined_urgency_viz = df_combined_urgency_viz.sort_values(by='Weighted Mean', ascending=False)

# Filter to the requested countries only
country_list = [
    'Argentina',
    'Australia',
    'Canada',
    'China',
    'Colombia',
    'France',
    'Germany',
    'Guatemala',
    'Paraguay',
    'Republic of Korea (the)',
    'Romania',
    'Russian Federation (the)',
    'Spain',
    'United Kingdom of Great Britain and Northern Ireland (the)',
    'United States of America (the)'
]

# Keep only those countries that exist in the data
df_combined_urgency_viz = df_combined_urgency_viz[df_combined_urgency_viz['Country'].isin(country_list)]

# Re-sort so bars are ordered by Weighted Mean descending
df_combined_urgency_viz = df_combined_urgency_viz.sort_values(by='Weighted Mean', ascending=False)

# Display the head of the processed (and filtered) data to show the combined urgency
display(df_combined_urgency_viz.head(20))

# Create the visualization
plt.figure(figsize=(12, 8))
sns.barplot(
    x='Weighted Mean',
    y='Country',
    data=df_combined_urgency_viz,
    palette='viridis'
)
plt.title('Total Urgency for Renewable Energy Transition by Country (Selected Countries)', fontsize=14)
plt.xlabel('Total Urgency (Combined %)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Keep only those countries that exist in the data
df_combined_urgency_viz = df_combined_urgency_viz[df_combined_urgency_viz['Country'].isin(country_list)]

# Re-sort so bars are ordered by Weighted Mean descending
df_combined_urgency_viz = df_combined_urgency_viz.sort_values(by='Weighted Mean', ascending=False)

# Display the head of the processed (and filtered) data to show the combined urgency
display(df_combined_urgency_viz.head(20))

# Create the visualization with red-to-green gradient
plt.figure(figsize=(12, 8))

# Normalize Weighted Mean to 0-1 range for color mapping
norm = plt.Normalize(vmin=df_combined_urgency_viz['Weighted Mean'].min(), 
                     vmax=df_combined_urgency_viz['Weighted Mean'].max())
colors = plt.cm.RdYlGn(norm(df_combined_urgency_viz['Weighted Mean'].values))

sns.barplot(
    x='Weighted Mean',
    y='Country',
    data=df_combined_urgency_viz,
    palette=colors
)
plt.title('Total Urgency for Renewable Energy Transition by Country (Selected Countries)', fontsize=14)
plt.xlabel('Total Urgency (Combined %)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Create chloropleth
import plotly.express as px

# Custom red → green gradient
red_green = [
    [0.0, 'red'],
    [0.2, '#FF0000'],   # light green
    [0.4, '#FF5500'],   # soft green
    [0.6, '#FFAA00'],   # medium green
    [0.8, '#AADD00'],   # strong green
    [1.0, '#00FF00']    # deep green
]

fig = px.choropleth(
    df_combined_urgency_viz,
    locations='Country',
    locationmode='country names',
    color='Weighted Mean',
    color_continuous_scale=red_green,
    # title='Support for Replacing Fossil Fuels with Renewable Energy (Selected Countries)',
    labels={'Percentage': 'Weighted Mean (%)'}
)

fig.update_layout(
    geo=dict(showframe=False, showcoastlines=True),
    coloraxis_colorbar=dict(title="Support (%)"),
    width=1400,    # increase figure width
    height=900,    # increase figure height
    margin=dict(l=10, r=10, t=80, b=10)  # tighten margins
)

fig.show()
```
